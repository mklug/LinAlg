This script is mainly to get you up and running.  You can play around with hitting a point over and over again with a rotation matrix.  If the angle of rotation (`theta` below) is not a rational multiple of $\pi$, then the point, when repeatedly rotated appears to fill up the unit circle. Can you prove this?  Can you quantify the rate at which the point and all of its rotations become dense in the unit circle?  What happens when the angle of rotation is a rational mutiple of $\pi$?

```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos
```

```python
pi = np.pi
# Random number between 0 and 1.  
epsilon = np.random.rand()
#theta = pi / 4
theta = 2*pi*epsilon
#theta = pi / 8 + 0.000001
```

```python
# Rotation matrix.
R = np.array([[cos(theta), (-1)*sin(theta)],
               [sin(theta), cos(theta)]])
v = np.array([1,0])
# Numpy notation for matrix multiplication.  
# R@v
applyR = lambda w: R@w; 
```

```python
points = [v]
# Around 250 starts to fill out the whole thing.  
iterations = 10000
for i in range(iterations): points.append(applyR(points[-1]))
xs, ys = [p[0] for p in points], [p[1] for p in points] 
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(xs, ys, edgecolor='k')
```
